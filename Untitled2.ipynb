# **Memuat dan Memproses Data dengan Gabor Filter**

```python
IMAGE_WIDTH = 224
IMAGE_HEIGHT = 224
IMAGE_SIZE = (IMAGE_WIDTH, IMAGE_HEIGHT)
IMAGE_CHANNELS = 3

# path = "/content/drive/MyDrive/Split_data/suku9/train"
path = "/content/drive/MyDrive/suku12/train"
lists_dir = os.listdir(path)
categories = []
file_name = []
for folder in lists_dir:
    files = os.listdir(path + "/" + folder)
    for file in files:
        path_file = path + "/" + folder + "/" + file
        file_name.append(path_file)
        categories.append(folder)

df = {"image": file_name, "categories": categories}
data = pd.DataFrame(df, index=None)
data = data.sample(frac=1)
data.head(5)

# Precompute Gabor Filter and save images
gabor_images_path = '/content/gabor_images/train/'
os.makedirs(gabor_images_path, exist_ok=True)

def apply_gabor_filter(image):
    kernels = []
    ksize = 31
    for theta in np.arange(0, np.pi, np.pi / 4):
        kernel = cv2.getGaborKernel((ksize, ksize), 4.0, theta, 10.0, 0.5, 0, ktype=cv2.CV_32F)
        kernels.append(kernel)
    accum = np.zeros_like(image)
    for kernel in kernels:
        fimg = cv2.filter2D(image, cv2.CV_8UC3, kernel)
        np.maximum(accum, fimg, accum)
    return accum

gabor_file_paths = []
for i, row in data.iterrows():
    img = cv2.imread(row['image'])
    img_gabor = apply_gabor_filter(img)
    new_path = os.path.join(gabor_images_path, os.path.basename(row['image']))
    cv2.imwrite(new_path, img_gabor)
    gabor_file_paths.append(new_path)

data['gabor_image'] = gabor_file_paths

# path_datatest = "/content/drive/MyDrive/Split_data/suku9/test"
path_datatest = "/content/drive/MyDrive/suku12/test"
lists_dir_datatest = os.listdir(path_datatest)
categories_datatest = []
file_name_datatest = []
for folder in lists_dir_datatest:
    files = os.listdir(path_datatest + "/" + folder)
    for file in files:
        path_file = path_datatest + "/" + folder + "/" + file
        file_name_datatest.append(path_file)
        categories_datatest.append(folder)

df_datatest = {"image": file_name_datatest, "categories": categories_datatest}
data_datatest = pd.DataFrame(df_datatest, index=None)
data_datatest = data_datatest.sample(frac=1)
nb_samples = data_datatest.shape[0]
data_datatest

# Precompute Gabor Filter and save images for test set
gabor_images_test_path = '/content/gabor_images/test/'
os.makedirs(gabor_images_test_path, exist_ok=True)

gabor_test_file_paths = []
for i, row in data_datatest.iterrows():
    img = cv2.imread(row['image'])
    img_gabor = apply_gabor_filter(img)
    new_path = os.path.join(gabor_images_test_path, os.path.basename(row['image']))
    cv2.imwrite(new_path, img_gabor)
    gabor_test_file_paths.append(new_path)

data_datatest['gabor_image'] = gabor_test_file_paths

data['categories'].value_counts().plot.bar()

total_dataset = len(data_datatest) + len(data)
print("Total dataset Suku: {} Suku".format(total_dataset))

data["categories"] = data["categories"].replace({0: 'moi', 1: 'nonmoi'})

train_df, validate_df = train_test_split(data, test_size=0.10, random_state=20)
train_df = train_df.reset_index(drop=True)
validate_df = validate_df.reset_index(drop=True)

train_df['categories'].value_counts().plot.bar()
validate_df['categories'].value_counts().plot.bar()
```

# **Mempersiapkan ImageDataGenerator tanpa Gabor Filter**

```python
batch_size = 64

train_datagen = ImageDataGenerator(
    rotation_range=30,
    rescale=1./255,
    shear_range=0.1,
    vertical_flip=True,
    zoom_range=0.2,
)

train_generator = train_datagen.flow_from_dataframe(
    dataframe=train_df,
    directory=None,  # Specify directory as None because full paths are provided in 'gabor_image'
    x_col='gabor_image',
    y_col='categories',
    target_size=IMAGE_SIZE,
    class_mode='categorical',
    batch_size=batch_size
)

validation_datagen = ImageDataGenerator(rescale=1./255)

validation_generator = validation_datagen.flow_from_dataframe(
    dataframe=validate_df,
    directory=None,  # Specify directory as None because full paths are provided in 'gabor_image'
    x_col='gabor_image',
    y_col='categories',
    target_size=IMAGE_SIZE,
    class_mode='categorical',
    batch_size=batch_size
)
```

# **Visualisasi Data**

```python
example_df = train_df.sample(n=1).reset_index(drop=True)
example_generator = train_datagen.flow_from_dataframe(
    dataframe=example_df,
    directory=None,  # Specify directory as None because full paths are provided in 'gabor_image'
    x_col='gabor_image',
    y_col='categories',
    target_size=IMAGE_SIZE,
    class_mode='categorical'
)

plt.figure(figsize=(12, 12))
for i in range(0, 10):
    plt.subplot(5, 3, i+1)
    for X_batch, Y_batch in example_generator:
        image = X_batch[0]
        plt.imshow(image)
        break
plt.tight_layout()
plt.show()
```

# **Membuat dan Melatih Model**

```python
base_model = EfficientNetB1(input_shape=(224, 224, 3),
                            include_top=False,
                            weights='noisy-student')

for layer in base_model.layers:
    layer.trainable = False

model = Sequential()
model.add(base_model)
model.add(Conv2D(640, kernel_size=(1, 1)))
model.add(Flatten())
model.add(Dense(320, kernel_initializer='he_uniform'))
model.add(BatchNormalization())
model.add(tf.keras.layers.ReLU())
model.add(Dropout(0.2))
model.add(Dense(2, activation='softmax'))

model.compile("adam", loss="binary_crossentropy", metrics=["accuracy"])
model.summary()

epochs = 30
history = model.fit(
    train_generator,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=1
)
```

# **Evaluasi Model**

```python
plt.plot(history.history["accuracy"], label="Akurasi Pelatihan")
plt.plot(history.history["val_accuracy"], label="Validasi Akurasi")
plt.legend()
plt.show()

plt.plot(history.history["loss"], label="Kesalahan Pelatihan")
plt.plot(history.history["val_loss"], label="Validasi Kesalahan")
plt.legend()
plt.show()

test_gen = ImageDataGenerator(rescale=1./255)
test_generator = test_gen.flow_from_dataframe(
    dataframe=data_datatest,
    directory=None,  # Specify directory as None because full paths are provided in 'gabor_image'
    x_col='gabor_image',
    y_col=None,
    class_mode=None,
    target_size=IMAGE_SIZE,
    batch_size=batch_size,
    shuffle=False
)

predict = model.predict(test_generator, steps=np.ceil(nb_samples / batch_size))

predicted_classes = np.argmax(predict, axis=1)
true_classes = np.array(data_datatest['categories'].replace(['moi', 'nonmoi'], [0, 1]))
class_labels = ['0', '1']

cm = confusion_matrix(true_classes, predicted_classes)
cm_df = pd.DataFrame(cm, index=['moi', 'nonmoi'], columns=['moi', 'nonmoi'])

plt.figure(figsize=(8, 8))
sns.heatmap(cm_df, annot=True)
plt.title('CNN EfficientNet-B1 \nAccuracy:{0:.3f}'.format(accuracy_score(true_classes, predicted_classes)))
plt.ylabel('Label Sebenarnya')
plt.xlabel('Label Prediksi')
plt.show()

report = metrics.classification_report(true_classes, predicted_classes, target_names=class_labels)
print(report)

model.save('geeksforgeeks.h5')
```
